```python
pkls = [f for f in os.listdir() if f.endswith('.pkl')]
```

```python
pkls = ['loss_l27b.pkl', 'loss_l38b.pkl',
        'loss_llamat27b_matsci.pkl', 'loss_llamat_38b_matsci.pkl',
        'loss_l27b_matsci.pkl', 'loss_l38b_matsci.pkl',
        'loss_llamat27b.pkl', 'loss_llamat_38b.pkl',
 ]

pname = ['llama-2-7b-eng','llama-3-8b-eng',
         'llamat-2-7b-matsci', 'llamat-3-8b-matsci',
         'llama-2-7b-matsci', 'llama-3-8b-matsci', 
         'llamat-2-7b-eng','llamat-3-8b-eng']
```

```python
for ff, fname in zip(pname, pkls):
    data = pickle.load(open(fname,'rb'))
    plt.figure(figsize=(8, 6))
    plt.contourf(data[0], data[1], data[2], levels=20, cmap="viridis",vmin=data[2].min(),vmax=data[2].max())
    plt.colorbar(label="Loss")
    plt.xlabel("Direction 1")
    plt.ylabel("Direction 2")
    plt.title(f"{ff}")
    plt.savefig(f'{ff}.png',dpi=600,bbox_inches='tight')
    plt.show()
```

```python
for ff, fname in zip(pname, pkls):
    data = pickle.load(open(fname, 'rb'))
    normalized_data = (data[2] - data[2].min()) / (data[2].max() - data[2].min())  # Normalize
    plt.figure(figsize=(8, 6))
    plt.contourf(data[0], data[1], normalized_data, levels=20, cmap="viridis")
    plt.colorbar(label="Normalized Loss")
    plt.xlabel("Direction 1")
    plt.ylabel("Direction 2")
    plt.title(f"{ff}")
    plt.savefig(f'norm_{ff}.png',dpi=600,bbox_inches='tight')
    plt.show()
```
