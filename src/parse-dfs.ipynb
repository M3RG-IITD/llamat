```python
def parse_checkpoint(cp):
    scores = {('qna', 'squadv2'): [],
     ('sar', 'synthesis_actions'): [],
     ('sc', 'sofc_sent'): [],
     ('re', 'structured_re'): [],
     ('re', 'sc_comics'): [],
     ('ee', 'sc_comics'): [],
     ('pc', 'glass_non_glass'): []}
    
    a = [i.split(' ') for i in cp.split('\n')]
    a_ = []
    for i in a:
        i_ = []
        for j in i:
           if j != '':
               i_.append(j)
        if len(i_) != 0:
            a_.append(i_)

    scores[('qna', 'squadv2')] = {'em' : a_[2][1], 'f1' : a_[3][1]}
    scores[('sar', 'synthesis_actions')] = {'micro_f1' : a_[4][2], 'macro_f1' : a_[5][2], 'ans_not_found' : a_[6][2]}
    scores[('sc', 'sofc_sent')] = {'micro_f1' : a_[4][3], 'macro_f1' : a_[5][3], 'ans_not_found' : a_[6][3]}
    scores[('re', 'structured_re')] = {'micro_f1' : a_[4][4], 'macro_f1' : a_[5][4], 'ans_not_found' : a_[6][4]}
    scores[('re', 'sc_comics')] = {'micro_f1' : a_[4][5], 'macro_f1' : a_[5][5], 'ans_not_found' : a_[6][5]}
    scores[('ee', 'sc_comics')] = {'micro_f1' : a_[4][6], 'macro_f1' : a_[5][6], 'ans_not_found' : a_[6][6]}
    scores[('pc', 'glass_non_glass')] = {'micro_f1' : a_[4][7], 'macro_f1' : a_[5][7], 'ans_not_found' : a_[6][7]}

    def floatify(d):
        for key, value in d.items():
            if isinstance(value, dict):  # If the value is a dictionary, apply recursion
                floatify(value)
            else:
                try:
                    d[key] = float(value)  # Try to convert the value to an integer
                except Exception as e:
                    print(e)
        return d
    scores = floatify(scores)
    return scores
```
