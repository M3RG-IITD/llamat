```python
import time
import concurrent.futures
from tqdm import tqdm
import json

def generate_single_request(prompt, model_name, max_tokens=1024, temperature=0.01, top_p=0.95, timeout=30):
    """
    Generate text for a single prompt using vLLM API
    """
    payload = {
        "model": model_name,
        "prompt": prompt + "output-",
        "max_tokens": max_tokens,
        "temperature": temperature,
        "top_p": top_p
    }
    
    try:
        response = requests.post(url, json=payload, timeout=timeout)
        response.raise_for_status()
        result = response.json()
        return result["choices"][0]["text"].strip()
    except Exception as e:
        print(f"Error generating for prompt: {str(e)}")
        return None

def batch_generate_sequential(df, model_name, max_tokens=1024, temperature=0.01, top_p=0.95, 
                            save_freq=100, output_file="generated_results.csv"):
    """
    Generate text for all rows sequentially with progress tracking
    """
    print(f"Starting sequential generation for {len(df)} rows...")
    
    generated_texts = []
    failed_indices = []
    
    for i, row in tqdm(df.iterrows(), total=len(df), desc="Generating"):
        prompt = row['prompt']
        generated_text = generate_single_request(prompt, model_name, max_tokens, temperature, top_p)
        
        if generated_text is not None:
            generated_texts.append(generated_text)
        else:
            generated_texts.append("")  # Empty string for failed generations
            failed_indices.append(i)
        
        # Periodic saving
        if (i + 1) % save_freq == 0:
            df_temp = df.copy()
            df_temp['generated_text'] = generated_texts + [""] * (len(df) - len(generated_texts))
            df_temp.to_csv(f"{output_file}_checkpoint_{i+1}.csv", index=False)
            print(f"Checkpoint saved at row {i+1}")
    
    # Add generated text to dataframe
    df['generated_text'] = generated_texts
    df.to_csv(output_file, index=False)
    
    print(f"Generation complete! Saved to {output_file}")
    print(f"Failed generations: {len(failed_indices)}")
    
    return df

def batch_generate_parallel(df, model_name, max_workers=4, max_tokens=1024, temperature=0.01, 
                          top_p=0.95, save_freq=100, output_file="generated_results_parallel.csv"):
    """
    Generate text for all rows in parallel with progress tracking
    """
    print(f"Starting parallel generation for {len(df)} rows with {max_workers} workers...")
    
    # Prepare prompts
    prompts = df['prompt'].tolist()
    
    generated_texts = [""] * len(df)
    failed_indices = []
    
    def process_batch(batch_indices):
        """Process a batch of indices"""
        batch_results = []
        for idx in batch_indices:
            prompt = prompts[idx]
            generated_text = generate_single_request(prompt, model_name, max_tokens, temperature, top_p)
            batch_results.append((idx, generated_text))
        return batch_results
    
    # Process in batches to avoid overwhelming the server
    batch_size = max(1, max_workers * 2)  # Smaller batches for better control
    
    with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers) as executor:
        # Submit batches
        futures = []
        for i in range(0, len(df), batch_size):
            batch_indices = list(range(i, min(i + batch_size, len(df))))
            future = executor.submit(process_batch, batch_indices)
            futures.append(future)
        
        # Collect results as they complete
        completed = 0
        for future in tqdm(concurrent.futures.as_completed(futures), 
                          total=len(futures), desc="Processing batches"):
            batch_results = future.result()
            
            for idx, generated_text in batch_results:
                if generated_text is not None:
                    generated_texts[idx] = generated_text
                else:
                    failed_indices.append(idx)
                
                completed += 1
                
                # Periodic saving
                if completed % save_freq == 0:
                    df_temp = df.copy()
                    df_temp['generated_text'] = generated_texts
                    df_temp.to_csv(f"{output_file}_checkpoint_{completed}.csv", index=False)
                    print(f"Checkpoint saved at {completed} completed generations")
    
    # Add generated text to dataframe
    df['generated_text'] = generated_texts
    df.to_csv(output_file, index=False)
    
    print(f"Parallel generation complete! Saved to {output_file}")
    print(f"Failed generations: {len(failed_indices)}")
    
    return df

# Example usage functions
def run_sequential_generation():
    """Run sequential generation for all rows"""
    return batch_generate_sequential(
        df[1800:], 
        model_names['llamat-2-cif'],
        max_tokens=1024,
        temperature=0.01,
        top_p=0.95,
        save_freq=50,
        output_file="llamat2_sequential_generated.csv"
    )

def run_parallel_generation(max_workers=2):
    """Run parallel generation for all rows"""
    return batch_generate_parallel(
        df, 
        model_names['llamat-2-cif'],
        max_workers=max_workers,
        max_tokens=1024,
        temperature=0.01,
        top_p=0.95,
        save_freq=50,
        output_file="llamat2_parallel_generated.csv"
    )
```
